## Aplicação das Features
Transformando dados brutos em variáveis úteis para o modelo

```python
# Importação de bibliotecas
import sys
import os

sys.path.append(os.path.abspath(".."))

import pandas as pd
from src.features.clean_text import limpeza_texto

# Definição do caminho do dataset
dataset_csv_path = '../data/dataset.csv'

# Carregamento do dataset
df = pd.read_csv(dataset_csv_path)
```

```python
# Aplicando função de limpeza 'limpeza_texto'
df['texto_limpo'] = df['texto_review'].apply(limpeza_texto)

print("Resultado após realização da limpeza de texto: ")
df.head()
```

```python
# Mapeamento para transformar dados de 'sentimento' em valores números 0 e 1
df['sentimento_value'] = df['sentimento'].map({'positivo': 1, 'negativo': 0})
df.head()
```

```python
print("DataFrame limpo após limpeza e mapeamento:\n")
df[['texto_limpo', 'sentimento_value']].sample(5)
```
